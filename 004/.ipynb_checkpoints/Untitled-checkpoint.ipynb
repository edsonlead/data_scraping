```python
import requests
import numpy as np
from bs4 import BeautifulSoup as bs
```

```python
URL = "https://saikoanimes.com/legendados/?fwp_paged=%i"
PAGES = range(1,10)
FILE = "list_anime.html"
BOOTSTRAP = "<link rel='stylesheet' href='https://maxcdn.bootstrapcdn.com/bootstrap/4.0.0-beta.2/css/bootstrap.min.css'>"
CHARSET = "<meta charset='ISO-8859-1'>"
TITLE = "Lista Animes do Site Saikô"
```

```python
print(PAGES)
PAGES
```

```python
for i in PAGES:
    url = URL%(i)
    req = requests.get(url)
    html = bs(req.content, "html.parser")
    data = html.find("div",{"class":"anime"})
    name_anime = data.find("div",{"class":"listtittle"}).get_text().strip()
    url_anime = data.find("a").get('href')
    #data = data.get_text().strip()
    print("PAGE {}".format(i))
    print(name_anime)
    print(url_anime)
```
