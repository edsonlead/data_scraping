```python
import re
import requests
import numpy as np
from bs4 import BeautifulSoup
```

```python
PAGE = range(1, 261)
URL =  "http://www.portaldatransparencia.gov.br/servidores/OrgaoExercicio-ListaServidores.asp?CodOrg=25201&Pagina=%i"
```

```python
test = range(1, 11)

for i in test:
    url = URL %(i)
    print(url)
    rq = requests.get(url)
    bsObj = BeautifulSoup(rq.content, "html.parser")
    
    detalha_servidor = bsObj.find("table", {"summary":"Lista de servidores por Nível da Função"}).find(href=re.compile("IdServidor="))
    detalha_servidor  = str(detalha_servidor).replace("<a href=\"","")
    detalha_servidor  = str(detalha_servidor).replace("</a>","")
    detalha_servidor  = re.sub(r"\">[A-Z]+( [A-Z]*)+", "", detalha_servidor )
    detalha_servidor  = str(detalha_servidor).replace("amp;","")
    print (detalha_servidor)
    
    url2 = "http://www.portaldatransparencia.gov.br/servidores/"
    url2 = url2+detalha_servidor
    #print(url2)
    rq2 = requests.get(url2)
    bsObj2 = BeautifulSoup(rq2.content, "html.parser")
    link_remuneracao = bsObj2.find("a", {"title":"Remuneração individual do servidor"})
    link_remuneracao = str(link_remuneracao).replace("<a href=\"","")
    link_remuneracao = str(link_remuneracao).replace("title=\"Remuneração individual do servidor\">","")
    link_remuneracao = str(link_remuneracao).replace("<img alt=\"Remuneração individual do servidor\" class=\"\" src=\"../images/botao-remuneracao4.jpg\"/>","")
    link_remuneracao = str(link_remuneracao).replace("\"","")
    link_remuneracao = str(link_remuneracao).replace("</a>","")
    link_remuneracao = str(link_remuneracao).replace("amp;","")
    link_remuneracao = str(link_remuneracao).replace(" ","")
    
    cargo = bsObj2.find_all("td", {"class":"tituloDetalhe"})
    #print(cargo)
    cargo = cargo[1]
    cargo = cargo.get_text()
    if (cargo == "Cargo Emprego:"):
        print(cargo)
    
    url3 = "http://www.portaldatransparencia.gov.br"
    url3 = url3+link_remuneracao
    #print(url3)
    rq3 = requests.get(url3)
    bsObj3 = BeautifulSoup(rq3.content, "html.parser")

    teste = bsObj3.find_all("td", {"class":"colunaValor"})
    nome = teste[0]
    nome = nome.get_text().strip(' \t\n\r')
    #nome = str(nome).replace("  ","")
    cpf = teste[1]
    cpf = cpf.get_text().strip(' \t\n\r')
    #cpf = str(cpf).replace(" ","")
    valor_bruto = teste[7]
    valor_bruto = valor_bruto.get_text()
    valor_liquido = teste[13]
    valor_liquido = valor_liquido.get_text()
    print(nome)
    print(cpf)
    print(valor_bruto)
    print(valor_liquido)
    #print(teste)
```
